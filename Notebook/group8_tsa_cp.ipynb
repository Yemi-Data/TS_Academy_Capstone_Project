## **Decomposition**

```python
# Decompose the avr_price to show: trend, seasonality, and white noise (residues)
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df['avr_price'], model='multiplicative', period=12)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

fig, axes = plt.subplots(4, 1, figsize=(10, 8))

# Original Time-Series
axes[0].plot(df['avr_price'], label='Original', color='blue')
axes[0].set_ylabel('Value', fontsize=9)
axes[0].legend(loc='upper left')
axes[0].grid(True, alpha=0.3)

# Trend: the long-term direction of the data
axes[1].plot(trend, label='Trend', color='darkred')
axes[1].set_ylabel('Value', fontsize=9)
axes[1].legend(loc='upper left')
axes[1].grid(True, alpha=0.3)

# Seasonality: this shows repeating patterns over fixed periods (daily, monthly, quarterly, yearly)
axes[2].plot(seasonal, label='Seasonality', color='darkgreen')
axes[2].set_ylabel('Value', fontsize=9)
axes[2].legend(loc='upper left')
axes[2].grid(True, alpha=0.3)

# White Noise/Residual:  the random fluctuation left after removing trend and seasonality
axes[3].plot(residual, label='Residual / Noise', color='gray')
axes[3].set_ylabel('Value', fontsize=9)
axes[3].legend(loc='upper left')
axes[3].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## **Stationarity Testing**

```python
# Create Stationarity test function using the ADF & KPSS tests
from statsmodels.tsa.stattools import adfuller, kpss

def test_stationarity(series):

    ''' This function tests for stationarity in a time series.
        It makes use of the Augmented Dickey-Fuller (ADF) to test for unit roots (non-stationary)
        and the Kwiatkowski–Phillips–Schmidt–Shin (KPSS) to test  if a time series is stationary around a mean or trend'''

    print("--- ADF Test Results ---")
    result_adf = adfuller(series)
    print(f'ADF Statistic: {result_adf[0]:.4f}')
    print(f'p-value: {result_adf[1]:.4f}')
    if result_adf[1] < 0.05:
        print("Conclusion: Reject H0, series is likely stationary (no unit root).")
    else:
        print("Conclusion: Fail to reject H0, series is likely non-stationary (has a unit root).")

    print("\n--- KPSS Test Results ---")
    result_kpss = kpss(series, regression='c', nlags='auto')
    print(f'KPSS Statistic: {result_kpss[0]:.4f}')
    print(f'p-value: {result_kpss[1]:.4f}')
    if result_kpss[1] < 0.05:
        print("Conclusion: Reject H0, series is likely non-stationary (not stationary around a deterministic trend).")
    else:
        print("Conclusion: Fail to reject H0, series is likely stationary (stationary around a deterministic trend).")

    print("\n--- Combined Test Conclusions ---")
    # Interpret combined results
    adf_stationary = result_adf[1] < 0.05
    kpss_stationary = result_kpss[1] >= 0.05

    if adf_stationary and kpss_stationary:
        print("The series is likely stationary (both tests agree).")
    elif not adf_stationary and not kpss_stationary:
        print("The series is likely non-stationary (both tests agree).")
    elif adf_stationary and not kpss_stationary:
        print("Conflicting results: ADF suggests stationarity, KPSS suggests non-stationarity.")
    elif not adf_stationary and kpss_stationary:
        print("Conflicting results: ADF suggests non-stationarity, KPSS suggests stationarity.")
    print("\n---------------------------------")
```

- This function tests for stationarity in a time series

- It makes use of the Augmented Dickey-Fuller `(ADF)` to test for **unit roots** (non-stationary)

- And the Kwiatkowski–Phillips–Schmidt–Shin `(KPSS)` to test  if a time series is stationary around a mean or trend

- If **ADF p-value < 0.05**, the data has no unit root `(stationary)`

- If **KPSS p-value >= 0.05**, the data is `stationary` around a trend

```python
# Stationarity test of original data
test_stationarity(df['avr_price'])
```

## **ACF (Autocorrelation) and PACF (Partial Autocorrelation) Plots**

```python
# Create function to make acf & pacf plots for selecting model parameters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

def plot_acf_pacf(series):

  fig, axes = plt.subplots(2, 1, figsize=(10, 8))

  plot_acf(series, lags=36, ax=axes[0]) # lags=36 shows 3 years
  axes[0].set_title('Autocorrelation Function (ACF)', fontsize=10, fontweight='bold')
  axes[0].set_ylabel('Correlation', fontsize=9)
  axes[0].grid(True, alpha=0.3)

  plot_pacf(series, lags=36, ax=axes[1])
  axes[1].set_title('Partial Autocorrelation Function (PACF)', fontsize=10, fontweight='bold')
  axes[1].set_ylabel('Correlation', fontsize=9)
  axes[1].grid(True, alpha=0.3)

  plt.tight_layout()
  plt.show()
```

```python
plot_acf_pacf(df['diff_price'].dropna())
```

- The `ACF` plot is used in selecting *Moving Average (MA)* Model Order `(q)`, usually using `k` if there's a sharp cut-off at *lag k*

- The `PACF` plot is used in selecting *Auto Regressive (AR)* Model Order `(p)`, usually using `k` if there's a sharp cut-off at *lag k*

- Both plots show sharp cut-off at *lag 1*. This gives 3 main model options:
  - ARIMA (1,1,0)
  - ARIMA (0,1,1)
  - ARIMA (1,1,1)

- We will try a naive baseline ARIMA model with order (p,d,q) = (1,1,1). Where `d` is the order of differencing done.

## **Train-test Split**

```python
# Train-test split with proper time-based splitting (not random)
train_size = int(len(df) * 0.8) # 80% train, 20% test (industry standard for train-test splits)

train = df['avr_price'][:train_size]
test = df['avr_price'][train_size:]
```

```python
# Visualize train and test data
plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.legend()
plt.title("Train–Test Split (Time-Based)")
plt.grid(True, alpha=0.3)
plt.show()
```

## **ARIMA Modelling & Evaluation**

```python
# Auto-Regressive (p) Integrated (d) Moving Average (q) Model

'''This model does not account for the seasonality of the data'''

from statsmodels.tsa.arima.model import ARIMA

train.index.freq = 'MS' # Specify index frequency
arima_model = ARIMA(train, order=(1, 1, 1))
fitted_arima_model = arima_model.fit()
```

```python
# Visualize ARIMA forecast prediction and compare to actual data
arima_forecast = fitted_arima_model.forecast(steps=len(test))

plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Actual')
plt.plot(arima_forecast, label='Forecast')
plt.legend()
plt.title("Forecast vs Actual")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Create model evaluation function
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error

def evaluate_model_performance(actual, predicted):

  mae = round((mean_absolute_error(actual, predicted)), 2)
  mape = round((mean_absolute_percentage_error(actual, predicted) * 100), 2)
  mse = round((mean_squared_error(actual, predicted)), 2)
  rmse = round((np.sqrt(mse)), 2)
  rsr = round((rmse / np.std(actual)), 2)

  metrics = [mae, mape, mse, rmse, rsr]

  return metrics
```

- Of these metrics, `2` will be most crucial for determining model performace:
  - `mape:` a value **<= 10%** will show a fairly good model
  - `rsr:` this shows the ratio `rmse:std` and a value **<= 0.7** will show good model performance

```python
# Evaluate ARIMA model performance
metrics = evaluate_model_performance(test, arima_forecast)

print(f'''
Mean Absolute Error (MAE): {metrics[0]}
Mean Absolute Percentage Error (MAPE): {metrics[1]}%
Mean Squared Error (MSE): {metrics[2]}
Root Mean Squared Error (RMSE): {metrics[3]}
RMSE-Observations Standard Deviation Ratio (RSR): {metrics[4]}
''')
```

- Both the `forecast` and the evaluation metrics show a **poor** performance of this ARIMA model.

- We will not bother optimizing the ARIMA `(p,d,q)` order. ARIMA fails to capture seasonal dynamics.

- For this, we will make use of the **Seasonal `(S)` ARIMA: SARIMA**

- Usually, for monthly data, like this one too, the *S-order is 12*.

- The `ACF` plot is used in determining the *Moving Average (MA)* Model Order `(Q)`, by selecting the number of significant **spikes** at _lags 12, 24, 36, ..._

- The `PACF` plot is used in determining the *Auto Regressive (AR)* Model Order `(P)`, by selecting the number of significant **spikes** at _lags 12, 24, 36, ..._

- ACF plot shows `1` significant spike at lag 12 and none after

- PACF plot shows significant spikes at lag 12 and 24 and none after

- We will try a naive baseline `SARIMA` model with order (p,d,q) x (P,D,Q)[S] = (1,1,1) x (2,1,1)[12]. Where `D` is the order of seasonal differencing done.

```python
# Run Baseline SARIMA model
from statsmodels.tsa.statespace.sarimax import SARIMAX

sarima_model = SARIMAX(
    train,
    order=(1,1,1),
    seasonal_order=(2,1,1,12)
)

fitted_sarima_model = sarima_model.fit()
```

```python
# Visualize SARIMA forecast prediction and compare to actual data
sarima_forecast = fitted_sarima_model.forecast(steps=len(test))

plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Actual')
plt.plot(sarima_forecast, label='Forecast')
plt.legend()
plt.title("Forecast vs Actual")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Evaluate SARIMA model performance
metrics = evaluate_model_performance(test, sarima_forecast)

print(f'''
Mean Absolute Error (MAE): {metrics[0]}
Mean Absolute Percentage Error (MAPE): {metrics[1]}%
Mean Squared Error (MSE): {metrics[2]}
Root Mean Squared Error (RMSE): {metrics[3]}
RMSE-Observations Standard Deviation Ratio (RSR): {metrics[4]}
''')
```

- Both the baseline SARIMA `forecast` and the evaluation metrics show a **good** performance for the model.

- The `mape (5.45%)` is significantly lower than the ARIMA model `(11.14%)` and than our threshold of `10%`

- The `rsr (0.51)` is also significantly lower than the ARIMA model `(1.12)` and than our threshold of `0.7`

- Though this baseline `SARIMA` model has performed well, we need to run an analysis of the model to know these metrics are significant and not by chance

- There are a lot of metrics here. But we will focus on `2:`
  - `P>|z|:` this confirms each parameter/order is statistically significant if the value for each is **< 0.05**
  - `Prob(Q):` this is the Ljung-Box test for the residuals and a value **>= 0.05** confirms the residuals as white noise

## **Systematic Model Comparison**

```python
# Run a grid search to identify best SARIMA model parameters based on AIC value

'''
The AIC/BIC values are used to penalize model complexity and high-order parameters over parsimony.
A lower value for the two means a better model. But balance is key
'''

# Import necessary libraries and packages
import statsmodels.api as sm
from itertools import product
import warnings

# Filter out specific warnings
warnings.filterwarnings("ignore", category=UserWarning, message="Non-invertible starting seasonal moving average")
warnings.filterwarnings("ignore", category=UserWarning, message="Non-stationary starting autoregressive parameters found.")
warnings.filterwarnings("ignore", category=UserWarning, message="Non-invertible starting MA parameters found.")
warnings.filterwarnings("ignore", category=sm.tools.sm_exceptions.ConvergenceWarning)

# Parameters to test
p = q = P = Q = range(0, 3) # This will test p,q,P,Q values from 0-2
pdq = list(product(p, [1], q)) # This creates all possible combinations for pdq keeping d=1
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(product(P, [1], Q))] # Keeps: D=1, S=12

# List to store results
results_list = []

for param in pdq:
    for seasonal_param in seasonal_pdq:
        try:
            # Attempt to fit model
            model = sm.tsa.statespace.SARIMAX(train, order=param, seasonal_order=seasonal_param)
            results = model.fit(disp=False)

            # Make forecast
            forecast = results.forecast(steps=len(test))

            # Evaluate model performance
            metrics = evaluate_model_performance(test, forecast)

            # Store results in a dictionary
            results_list.append({
                'param': param,
                'seasonal_param': seasonal_param,
                'aic': results.aic,
                'bic': results.bic,
                'mae': metrics[0],
                'mape': metrics[1],
                'mse': metrics[2],
                'rmse': metrics[3],
                'rsr': metrics[4]
            })
        except Exception as e:
            # Skip invalid parameter combinations and models that fail to fit
            continue

# Create DataFrame from results list
aic_results_df = pd.DataFrame(results_list)

# Sort by AIC and display the top 5
aic_results_df = aic_results_df.sort_values(by='aic').reset_index(drop=True)

print("--- Top 5 SARIMAX models by AIC: ---\n")
display(aic_results_df.head())
```

```python
# Create function to run & analyse top 5 SARIMA models
def run_sarima_model_and_summary(order_list, seasonal_order_list, train_data, test_data):
  # 1. Initialize a single plot for all forecasts
  plt.figure(figsize=(11, 5))
  plt.plot(test_data, label='Actual Test Data')

  for i in range(len(order_list)):
    order = order_list[i]
    seasonal_order = seasonal_order_list[i]
    print(f"\n--- Running SARIMA Model with order={order} and seasonal_order={seasonal_order} ---\n")

    # 2. Initialize and fit the SARIMAX model
    model = SARIMAX(
        train_data,
        order=order,
        seasonal_order=seasonal_order
    )

    fitted_model = model.fit(disp=False) # disp=False suppresses convergence messages
    print(fitted_model.summary())

    # 3. Generate forecasts
    forecast = fitted_model.forecast(steps=len(test_data))
    forecast.index = test_data.index # Ensure forecast index matches test_data index

    # 4. Plot current model's forecast on the same graph
    plt.plot(forecast, label=f'Forecast (SARIMA{order}x{seasonal_order})')

  # 5. Final plot settings after all models have been plotted
  plt.title('SARIMA Forecasts vs Actual')
  plt.xlabel('Date')
  plt.xticks(rotation=45)
  plt.ylabel('Price Index')
  plt.legend()
  plt.grid(True, alpha=0.3)
  plt.show()
```

### **Model Limitations**

1. **Forecast Horizon**: Model is most reliable for 1-3 month forecasts. Uncertainty increases significantly beyond 6 months.

2. **Structural Breaks**: Model assumes past patterns continue. Major policy changes, global crises, or structural reforms may require model retraining.

3. **External Shocks**: Unpredictable events (geopolitical crises, pandemics, oil price shocks) are not captured in historical patterns.
